# 学習用テキストデータ生成スクリプト

### ライブラリのインポート

```python
import glob
import time
```

### 前処理

```python
file_list = glob.glob('data/*')
# count = 0
text_dic_list = []
for file in file_list:
    print(file)
    with open(file, encoding="cp932") as f:
        line_count = 0
        text_list = []
        text_dic = {}
        lines = f.read().splitlines()
        for li, line in enumerate(lines):
            line = line.rstrip()
            if li == 0:
                text_dic['title'] = line
            if line[0:3] == '底本：':
                break
            if line_count==2:
                if line == '':
                    continue
                if line[0] == '［':
                    continue
                if "《" in line:
                    while "《" in line:
                        line = line[:line.find("《")]+line[line.find("》")+1:]
                text_list.append(line.replace('\u3000','')+'<|endoftext|>')
            if line[0:4] == '----':
                line_count+=1
        text_dic['text'] = text_list
        text_dic_list.append(text_dic)
        f.close()
    time.sleep(0.1)
    # count+=1
    # if count==200:
    #     # print(text_dic_list)
    #     break
```

### 前処理したテキストを保存

```python
for text_dic in text_dic_list:
    with open('output/'+text_dic['title']+'.txt', 'w', encoding='utf-8') as f:
        for text in text_dic['text']:
            f.write(text)
```
